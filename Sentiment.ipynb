```python
import pandas as pd
```

```python
#headerList = ['id','tweet'] #header=headerList
train = pd.read_csv("sentimental.csv",encoding='latin-1')
#train.columns=["a", "b"]
```

```python
train.info()
```

```python
train.head()
```

```python
train["word_count"] = train["tweet"].apply(lambda x:len(str(x).split(" ")))
train["char_count"]=train["tweet"].str.len()
train.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
train["stopwords"] =train ["tweet"].apply(lambda x: len([x for x in x.split()if x in stop]))
train["hashtag"] =train ["tweet"].apply(lambda x: len([x for x in x.split()if x.startswith("#")]))
train["upper"] =train ["tweet"].apply(lambda x: len([x for x in x.split()if x.isupper()]))
train.head()
```

```python
train["tweet_clean"]=train["tweet"].str.replace("[^\w\s]","")
```

```python
train.head()
```

```python
train['tweet_clean'] = train['tweet_clean'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['tweet_clean'] = train['tweet_clean'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
from textblob import TextBlob
train["tweet_clean"]  [:5].apply(lambda x: str(TextBlob(x).correct()))
```

```python
train["tweet_clean"] [:].apply(lambda x: str(TextBlob(x).sentiment))
```

```python
train['sentiment'] = train['tweet_clean'].apply(lambda x: TextBlob(x).sentiment[0] )
train.head(28)
```
